```python
# Loop over columns of df_test and df_train
for feature in df_test.columns:
    if feature in df_train.columns:
        # Align the two Series by their index before comparison
        if df_test[feature].equals(df_train[feature]):
            print(f"Column '{feature}' is identical in both df_test and df_train")
        else:
            print(f"Column '{feature}' is different in df_test and df_train")
    else:
        print(f"Column '{feature}' is not present in df_train")
```
