```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("Customer Call List.xlsx")
```

```python
for i in df.columns:
  print(i)
```

```python
df.shape
```

```python
df.size
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.size
```

```python
df = df.drop(columns = 'Not_Useful_Column')
```

```python
df.head()
```

```python
df['Last_Name']= df['Last_Name'].str.strip('/\-.,_')
```

```python
df
```

```python
print(df['Phone_Number'].dtype)
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]', '', regex = True)
```

```python
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'].dtype
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
```

```python
df
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('nan--', '', regex= True)
df['Phone_Number']= df['Phone_Number'].str.replace('--', '', regex= True)
```

```python
df
```

```python
df['Paying Customer']= df['Paying Customer'].replace('Yes', 'Y')
df['Paying Customer']= df['Paying Customer'].replace('No', 'N')
```

```python
df
```

```python
df['Do_Not_Contact']= df['Do_Not_Contact'].replace('Yes', 'Y')
df['Do_Not_Contact']= df['Do_Not_Contact'].replace('No', 'N')
df['Do_Not_Contact']= df['Do_Not_Contact'].fillna('Y')
```

```python
df
```

```python
df = df[df['Phone_Number'] != '']
```

```python
df
```

```python
# Drop rows where 'Do_Not_Contact' is 'Y'
df = df[df['Do_Not_Contact']!= 'Y']
```
